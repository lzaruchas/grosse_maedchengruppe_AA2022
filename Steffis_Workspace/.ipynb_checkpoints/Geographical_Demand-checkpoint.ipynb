```python
# libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'no-latex'])
```

```python
# import
la = pd.read_csv("prepared_data.csv", parse_dates=["start_time","end_time"],low_memory=False)
```

```python
# new data set: coordinates  
# https://gbfs.bcycle.com/bcycle_lametro/station_information.json
coordinates = pd.read_csv("station_information.csv")
coordinates.drop(columns=['ttl', 'data__stations__rental_uris__ios','data__stations__rental_uris__android','version','last_updated'])
coordinates.head()
```

```python
la.head()
```

```python
start_station = la['start_station_name'].value_counts()
end_station = la['end_station_name'].value_counts()
```

```python
start_station_df = start_station.to_frame()
start_station_df = start_station_df.rename(columns={'start_station_name': 'count'})
start_station_df['start_station_name'] = start_station_df.index
start_station_df
```

```python
end_station_df = end_station.to_frame()
end_station_df = end_station_df.rename(columns={'end_station_name': 'count'})
end_station_df['end_station_name'] = end_station_df.index
end_station_df
```

```python
fig, ax = plt.subplots(figsize=[15,5])
plt.setp(ax.xaxis.get_majorticklabels(),rotation=90,fontsize=10)
ax.bar(start_station_df['start_station_name'].iloc[:50], start_station_df['count'].iloc[:50])
```

```python
fig, ax = plt.subplots(figsize=[15,5])
plt.setp(ax.xaxis.get_majorticklabels(),rotation=90,fontsize=10)
ax.bar(start_station_df['start_station_name'].iloc[-50:], start_station_df['count'].iloc[-50:])
plt.show()
```
