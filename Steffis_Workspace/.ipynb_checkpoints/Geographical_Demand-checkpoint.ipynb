```python
# libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'no-latex'])
```

```python
# import
la = pd.read_csv("prepared_data.csv", parse_dates=["start_time","end_time"],low_memory=False)
```

```python
la.head()
```

```python
start_station = la['start_station_name'].value_counts()
end_station = la['end_station_name'].value_counts()
```

```python
start_station_df = start_station.to_frame()
start_station_df = start_station_df.rename(columns={'start_station_name': 'count'})
start_station_df['start_station_name'] = start_station_df.index
start_station_df
```

```python
fig, ax = plt.subplots(figsize=[15,5])
plt.setp(ax.xaxis.get_majorticklabels(),rotation=90,fontsize=10)
ax.bar(start_station_df['start_station_name'].iloc[:50], start_station_df['count'].iloc[:50])
```
