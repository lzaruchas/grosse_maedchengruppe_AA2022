# KPI utilization (number of rides)

```python
# libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
from statistics import mean
from datetime import date
```

## Data Preparation

```python
# import
# la = pd.read_csv("prepared_data.csv", parse_dates=["start_time","end_time"],low_memory=False)
# la_hourly =  pd.read_csv("datasets_merged_hour.csv",low_memory=False)
la_hourly =  pd.read_csv("prediction_data.csv",low_memory=False)
```

```python
la_hourly.head(2)
```

## Number of trips per hour

```python
# Overview trips per hour of year
plt.figure(figsize = (18,5))
plt.plot(la_hourly.index, la_hourly["trips"], label='Trips per hour')
plt.title('Absolute number of trips per hour')
plt.ylabel('Trips')
plt.xlabel('Hour')
plt.show()
# plt.savefig("Absolute number of trips per hour.png")
```

## Hours with the highest number of rides

```python
la_hourly = la_hourly.sort_values(by='trips', ascending=False)
la_hourly.head(5)
```

```python
# plot of the 25 hours with the highest number of rides
fig, ax = plt.subplots(figsize=[15,5])
plt.setp(ax.xaxis.get_majorticklabels(),rotation=90,fontsize=10)
ax.bar(la_hourly['start_time'].iloc[:25], la_hourly['trips'].iloc[:25])
plt.title('25 hours with the highest number of rides')
# plt.savefig("25 hours with the highest number of rides.png")
```
