#### 2. Trips based on precipitation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML
from IPython.display import display, HTML
display(HTML(""))
import scienceplots
plt.style.use(['science', 'no-latex'])
from statistics import mean
```

```python
# import prepared data set
df_la = pd.read_csv("prepared_data.csv", parse_dates=["start_time","end_time","Date"],low_memory=False)
df_weather = pd.read_csv("prepared_data_weather.csv", parse_dates=["date_time"],low_memory=False)
```

```python
df_la.head(3)
```

```python
df_weather.head(5)
```

```python
def get_trips(datetime):
    return df_la[(df_la["Month"] == datetime.month) & (df_la["Day"] == datetime.day) & (df_la["Hour"] == datetime.hour)].size
```

```python
df_weather["num_trips"] = df_weather["date_time"].apply(get_trips)
df_weather["hour"] = df_weather.index
```

```python
df_weather.head(5)
```

```python
# Overview trips per hour of year
plt.plot(df_weather.index, df_weather["num_trips"], label='Trips per hour')
plt.title('Absolute number of trips per hour')
plt.ylabel('Trips')
plt.xlabel('Hour')
plt.show()
```

```python
# Overview precipitation per hour of year
plt.plot(df_weather.index, df_weather["precip"], label='Precipitation per hour with 1 = precipitation and 0 = no precipitation')
plt.ylabel('Precipitation')
plt.xlabel('Hour')
plt.show()
```

```python
# number of days without and with precipitation
precip_count = df_weather["precip"].value_counts()
df = precip_count.to_frame().rename(columns = {'precip': 'count'})
df['precip'] = df.index
df.reset_index(drop=True, inplace=True) 
df
```

```python
# number of days without and with precipitation
fig, ax = plt.subplots()
plt.setp(ax.xaxis.get_majorticklabels(),rotation=90,fontsize=10)
ax.bar(df["precip"], df["count"])
plt.title('Trips based on precipitation with 0 = no precipitation and 1 = precipitation')
```

```python
# Mean of number of trips when precipitation
hours_with_precip = df_weather[df_weather["precip"] == 1]
mean(hours_with_precip["num_trips"])
```

```python
# Mean of numbers of trips when no precipitation
hours_without_precip = df_weather[df_weather["precip"] == 0]
mean(hours_without_precip["num_trips"])
```

```python
# number of rides per hour with and without precipitation
plt.figure(figsize = (18,4))
plt.scatter(hours_without_precip.index, hours_without_precip["num_trips"], color = 'g', s=10)
plt.scatter(hours_with_precip.index, hours_with_precip["num_trips"], color = 'b', s=10)
plt.xlabel("Number of trips")
plt.ylabel("Hour")
plt.title("Number of rides per hour with and without precipitation")
plt.show()
```

```python
# number of rides per hour and precipitation
fig, ax = plt.subplots(figsize=[18,4])
plt.setp(ax.xaxis.get_majorticklabels(),rotation=90,fontsize=10)
ax.bar(df_weather.index, df_weather["num_trips"])
ax.set_ylim([0,max(df_weather["num_trips"])])
plt.title('Precipitation hourly and number of trips per hour')
```

```python
plt.bar(df_weather.index, df_weather["num_trips"], tick_label = df_weather.index, color = ['red', 'green'])
```
