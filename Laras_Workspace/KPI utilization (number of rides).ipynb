```python
# libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
from statistics import mean
from datetime import date
```

## Data Preparation

```python
# import
la = pd.read_csv("prepared_data.csv", parse_dates=["start_time","end_time"],low_memory=False)
la_hourly =  pd.read_csv("datasets_merged_hour.csv",low_memory=False)
```

```python
la.head(2)
```

```python
la_hourly = la_hourly.drop(columns=["max_temp","min_temp","precip"])
la_hourly.head(2)
```

```python
# number of bikes
bikes = la['bike_id'].unique()
number_of_bikes = len(bikes)
number_of_bikes
```

## Number of trips per hour

```python
# Overview trips per hour of year
plt.figure(figsize = (18,5))
plt.plot(la_hourly.index, la_hourly["num_trips_hour"], label='Trips per hour')
plt.title('Absolute number of trips per hour')
plt.ylabel('Trips')
plt.xlabel('Hour')
plt.show()
```
